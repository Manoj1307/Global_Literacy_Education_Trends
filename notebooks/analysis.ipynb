# 2. Data Sourcing

## 2.1 Adult Literacy Rate

```python
url = "https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true"
storage_options = {'User-Agent': 'Mozilla/5.0'}
df_adultliteracy = pd.read_csv(url, storage_options=storage_options)
```

# 4. Data Cleaning

```python
df_literacy.head(5)
```

```python
df_literacy = df_literacy.drop(columns=['code_x', 'code_y'])
```

```python
df_literacy
```

```python
df_literacy.columns = ['ENTITY', 'YEAR', 'ADULT_LITERACY', 'YOUTH_LITERACY_M' ,'YOUTH_LITERACY_F', 'OWID_REGION']
```

```python
df_literacy[['ADULT_LITERACY', 'YOUTH_LITERACY_M', 'YOUTH_LITERACY_F', 'OWID_REGION']]= df_literacy.groupby('ENTITY')[['ADULT_LITERACY', 'YOUTH_LITERACY_M', 'YOUTH_LITERACY_F', 'OWID_REGION']].ffill()
```

```python
df_literacy[['ADULT_LITERACY', 'YOUTH_LITERACY_M', 'YOUTH_LITERACY_F', 'OWID_REGION']]= df_literacy.groupby('ENTITY')[['ADULT_LITERACY', 'YOUTH_LITERACY_M', 'YOUTH_LITERACY_F', 'OWID_REGION']].bfill()
```

```python
df_literacy.loc[df_literacy['ENTITY'] == 'World', 'OWID_REGION'] = 'World'
```

```python
df_literacy = df_literacy.dropna(subset=['OWID_REGION'])
```

```python
df_literacy['ADULT_LITERACY'] = df_literacy.groupby('OWID_REGION')['ADULT_LITERACY'].transform(lambda x: x.fillna(x.mean()))
```

```python
df_literacy['YOUTH_LITERACY_M'] = df_literacy.groupby('OWID_REGION')['YOUTH_LITERACY_M'].transform(lambda x: x.fillna(x.mean()))
```

```python
df_literacy.isnull().sum()
```

```python
df_literacy.to_csv("../data/df_literacy.csv", index = False)
```
