# Exploratory Data Analysis (EDA) - Plotly Version

This notebook analyzes the education and literacy datasets prepared during the feature engineering step using **Plotly** for interactive data visualizations.
1. Average Years of Education
2. Illiteracy Rates
3. Literacy Rates

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
# import warnings

# warnings.filterwarnings('ignore')
```

## 1. Load the Datasets

```python
# Define paths
path_avg = '../data/implemented_data/avg_years_of_education_p.csv'
path_ill = '../data/implemented_data/illeteracy_p.csv'
path_lit = '../data/implemented_data/literacy_p.csv'

# Load dataframes
df_avg = pd.read_csv(path_avg)
df_ill = pd.read_csv(path_ill)
df_lit = pd.read_csv(path_lit)

print("Data loaded successfully!")
```

## 2. Analysis: Average Years of Education (`df_edu`)

```python
df_avg.head()
```

```python
df_avg.describe()
```

### Univariate Analysis (df_edu)

```python
fig = make_subplots(rows=1, cols=3, subplot_titles=(
    'Literacy Distribution', 
    'AVG_YEARS_OF_EDUCATION Distribution', 
    'GDP_PER_CAPITA (Female) Distribution'))

fig.add_trace(go.Histogram(x=df_avg['LITERACY_RATE'], name='Literacy', marker_color='red', opacity=0.7), row=1, col=1)
fig.add_trace(go.Histogram(x=df_avg['AVG_YEARS_OF_EDUCATION'], name='Avg_years_of_education', marker_color='green', opacity=0.7), row=1, col=2)
fig.add_trace(go.Histogram(x=df_avg['GDP_PCAP'], name='GDP_per_capta', marker_color='blue', opacity=0.7), row=1, col=3)
fig.update_layout(title_text='Average years of education over GDP per capita Distributions', height=500, showlegend=True)
fig.show()
```

### Insights: Univariate Analysis (df_edu)
- **Average Years of Education**: Centered around 6.5 years globally, with a minimum of 4.2 and maximum of 9.5 years, showing significant disparity.
- **GDP Per Capita**: Highly skewed distribution. The median ($12,636) is much lower than the mean ($21,885) and the max ($174,569), highlighting global wealth inequalities.
- **Literacy Rate**: The median literacy rate is 82.2%, indicating that a majority of the global population has basic reading skills, though lower quartiles show room for improvement.

### Bivariate Analysis (df_edu)

```python
fig = px.scatter(df_avg, x='GDP_PCAP', y='AVG_YEARS_OF_EDUCATION', color='REGION', 
                 log_x=True, hover_name='COUNTRY', hover_data=['YEAR'],
                 title='GDP per Capita (Log Scale) vs Average Years of Education',
                 opacity=0.7)
fig.show()

fig = px.scatter(df_avg, x='LITERACY_RATE', y='AVG_YEARS_OF_EDUCATION', color='REGION',
                 hover_name='COUNTRY', hover_data=['YEAR'],
                 title='Literacy Rate vs Average Years of Education',
                 opacity=0.7)
fig.show()
```

```python
fig = px.scatter(df_avg, x='REGION', y='AVG_YEARS_OF_EDUCATION', color='REGION',
             title='Average Years of Education by Region')
fig.show()
```

```python
# Correlation Matrix
numeric_cols = df_avg.select_dtypes(include=[np.number]).columns
corr_matrix = df_avg[numeric_cols].corr()

fig = px.imshow(corr_matrix, text_auto=True, aspect="auto", 
                title='Correlation Matrix for Education Dataset',
                color_continuous_scale='RdBu_r')
fig.show()
```

### Insights: Bivariate Analysis (df_edu)
- **Education & Literacy**: There is a strong positive correlation (r=0.89) between Average Years of Education and Literacy Rate, confirming that longer schooling directly translates to better literacy.
- **Wealth & Education**: GDP per Capita shows a notable positive correlation (r=0.49) with education years. Wealthier regions tend to provide longer education.
- **Education Index**: Shows a near-perfect correlation (r=0.96) with the average years of education, validating its use as a primary educational metric.

## 3. Analysis: Illiteracy (`df_ill`)

```python
df_ill.head()
```

```python
df_ill.describe()
```

### Univariate Analysis (df_ill)

```python
fig1 = px.histogram(df_ill, x='ILLITERACY_RATE', nbins=30, marginal="violin",
                   title='Distribution of Illiteracy Rate', color_discrete_sequence=['purple'])
fig1.show()

fig2 = px.histogram(df_ill, x='ILLITERACY TO LITERACY RATIO', nbins=30, marginal="violin",
                   title='Distribution of Illiteracy:Literacy Ratio', color_discrete_sequence=['orange'])
fig2.show()
```

### Insights: Univariate Analysis (df_ill)
- **Illiteracy Rate**: Global average is ~17.9%, but the median is 9.0%. The right-skewed distribution indicates that while most countries have low illiteracy, a few still suffer from rates up to 89%.
- **Literacy Growth**: Positive growth (mean 3.06%) highlights successful global educational interventions over the years.
- **Illiteracy-to-Literacy Ratio**: Primarily concentrated near 0, but extends up to 8.09 in extreme cases.

### Bivariate Analysis (df_ill)

```python
fig = px.scatter(df_ill, x='LITERACY_RATE', y='ILLITERACY_RATE', 
                 hover_name='COUNTRY', hover_data=['YEAR'],
                 title='Literacy Rate vs Illiteracy Rate', 
                 color_discrete_sequence=['darkgreen'], opacity=0.6)
fig.show()
```

### Insights: Bivariate Analysis (df_ill)
- **Illiteracy vs Literacy**: A perfect negative correlation (r=-1.0), as expected, since they are largely inversely proportional components.
- **Growth Rate Trends**: Literacy growth rate shows a slight positive correlation with illiteracy rate (r=0.30) - suggesting countries with higher illiteracy have more room for rapid growth.
- **Ratio Dependencies**: The illiteracy-to-literacy ratio strongly mirrors the illiteracy rate itself (r=0.85).

## 4. Analysis: Literacy gaps and demographics (`df_lit`)

```python
df_lit.head()
```

```python
df_lit.describe()
```

### Univariate Analysis (df_lit)

```python
fig = make_subplots(rows=1, cols=3, subplot_titles=(
    'Adult Literacy Distribution', 
    'Youth Literacy (Male) Distribution', 
    'Youth Literacy (Female) Distribution'))

fig.add_trace(go.Histogram(x=df_lit['ADULT_LITERACY'], name='Adult', marker_color='blue', opacity=0.7), row=1, col=1)
fig.add_trace(go.Histogram(x=df_lit['YOUTH_LITERACY_M'], name='Youth Male', marker_color='cyan', opacity=0.7), row=1, col=2)
fig.add_trace(go.Histogram(x=df_lit['YOUTH_LITERACY_F'], name='Youth Female', marker_color='magenta', opacity=0.7), row=1, col=3)

fig.update_layout(title_text='Literacy Distributions', height=500, showlegend=False)
fig.show()
```

### Insights: Univariate Analysis (df_lit)
- **Youth vs Adult Literacy**: Youth literacy (median 97.6% for males, 98.0% for females) is significantly higher than adult literacy (median 90.0%), proving generational progress in education.
- **Gender Gap**: The average gender literacy gap is 4.8%. While the median is a low 1.0%, extreme cases reach up to a 52.2% gap.
- **Adult vs Youth Gap**: The mean gap of 8.4% confirms recent historical acceleration in youth education access.

### Bivariate Analysis (df_lit)

```python
fig = px.scatter(df_lit, x='YOUTH_LITERACY_M', y='YOUTH_LITERACY_F', color='REGION',
                 hover_name='COUNTRY', hover_data=['YEAR'],
                 title='Youth Literacy: Male vs Female', opacity=0.7)
# Add equality line
fig.add_shape(type='line', x0=0, y0=0, x1=100, y1=100, line=dict(color='red', dash='dash'))
fig.show()

fig = px.scatter(df_lit, x='YOUTH LITERACY AVERAGE', y='ADULT_LITERACY', color='REGION',
                 hover_name='COUNTRY', hover_data=['YEAR'],
                 title='Youth Literacy Avg vs Adult Literacy', opacity=0.7)
# Add equality line
fig.add_shape(type='line', x0=0, y0=0, x1=100, y1=100, line=dict(color='red', dash='dash'))
fig.show()
```

```python
fig = px.box(df_lit, x='REGION', y='LITERACY GAP GENDER', color='REGION',
             title='Gender Literacy Gap by Region (Male - Female %)')
fig.show()
```

### Insights: Bivariate Analysis (df_lit)
- **Gender Parity via Overall Education**: There is a strong negative correlation (r=-0.79) between Adult Literacy and Gender Gap. As general literacy increases, gender disparities tend to shrink.
- **Youth Trajectory**: Male and female youth literacy are highly correlated (r=0.93). A country improving male youth education typically improves female youth education synchronously.
- **Intergenerational Growth**: Adult literacy strongly predicts youth literacy (r=0.93), showing cumulative educational foundations over generations.
