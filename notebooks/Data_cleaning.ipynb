# Creating Data from for Implementation

## df_literacy(Merging adult_literacy and youth_literacy)

```python
df_lit = pd.merge(df_alit, df_ylit, on = ['country','code','year'], how = 'outer')
```

```python
df_lit.info()
```

```python
cols = ["adult_literacy", "youth_literacy_m", "youth_literacy_f"]
# ffill per country
df_lit[cols] = df_lit.groupby("country")[cols].transform(lambda x: x.ffill())
# bfill per country (catches remaining NAs at the start of a group)
df_lit[cols] = df_lit.groupby("country")[cols].transform(lambda x: x.bfill())
```

```python
df_lit[df_lit.isnull().any(axis=1)]
```

```python
# Step 1: Build a country -> region mapping from existing non-null values
region_map = df_lit.dropna(subset=["region"]).drop_duplicates("country").set_index("country")["region"]

# Step 2: Fill missing region using the mapping
df_lit["region"] = df_lit["region"].fillna(df_lit["country"].map(region_map))
```

```python
df_lit[cols] = df_lit.groupby("region")[cols].transform(lambda x: x.fillna(x.median()))
```

```python
df_lit.dropna(how="any", inplace=True)
```

```python
df_lit = df_lit[df_lit["year"] >= 1990]
```

```python
df_lit.reset_index(drop=True, inplace=True)
```

```python
df_lit.info()
```

```python
df_lit.head()
```

```python
# Convert all string values to uppercase
df_lit = df_lit.apply(lambda x: x.str.upper() if x.dtype == "str" else x)

# Convert column names to uppercase
df_lit.columns = df_lit.columns.str.upper()
```

```python
df_lit.duplicated().sum()
```

```python
df_lit.to_csv("../data/implemented_data/literacy_p.csv", index = False)
```

## df_gdp_schooling(Merging GDP per captia and average years of education)

```python
df_gsc = pd.merge(df_gc, df_sc, on = ['country','code','year','region'], how = 'outer')
```

```python
df_gsc.head()
```

```python
df_gsc = df_gsc[df_gsc["year"] >= 1990]
```

```python
df_gsc.info()
```

```python
df_gsc["gdp_pcap"] = df_gsc.groupby("country")["gdp_pcap"].transform(lambda x: x.ffill().bfill())
```

```python
df_gsc["gdp_pcap"] = df_gsc.groupby("region")["gdp_pcap"].transform(lambda x: x.ffill().bfill())
```

```python
df_gsc[["literacy_rate", "avg_years_of_education", "population"]] = df_gsc.groupby("country")[["literacy_rate", "avg_years_of_education", "population"]].transform(lambda x: x.ffill().bfill())
```

```python
df_gsc[["literacy_rate", "avg_years_of_education", "population"]] = df_gsc.groupby("region")[["literacy_rate", "avg_years_of_education", "population"]].transform(lambda x: x.mean())
```

```python
df_gsc.info()
```

```python
df_gsc.duplicated().sum()
```

```python
df_gsc.reset_index(drop=True, inplace=True)
```

```python
# Convert all string values to uppercase
df_gsc = df_gsc.apply(lambda x: x.str.upper() if x.dtype == "str" else x)

# Convert column names to uppercase
df_gsc.columns = df_gsc.columns.str.upper()
```

```python
df_gsc.to_csv("../data/implemented_data/avg_years_of_education_p.csv", index = False)
```
