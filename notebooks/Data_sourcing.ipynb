# Data Quality Assessment

## Adult Literacy Rate

```python
df_alit.head()
```

```python
df_alit[["adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99"]] = df_alit[["adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99"]].round(2)
```

```python
df_alit.isnull().sum()
```

```python
len(df_alit)
```

```python
# Percent of missing values
(df_alit.isnull().sum() / len(df_alit)) * 100
```

```python
df_alit.duplicated().sum()
```

```python
df_alit[df_alit.duplicated()]
```

```python
col = 'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99'
((df_alit[col] < 0) | (df_alit[col] > 100)).sum()
```

```python
now = dt.now()
(df_alit['year'] > now.year).sum()
```

**Assessment:** There are some missing values in the column **code**, **Columns** must be renamed and the groups, The **entity** columns has some grouped countries which needs to be removed
